```python
import pandas as pd
import numpy as np
```

# Load Data

```python
df = pd.read_csv("../../raw_data/NFL/NFL_scrapped.csv")
df.head()
```

## Format of the data:

- Wins, Losses and Ties per Team per Season from 2000 to 2021
- We only need Year, NFL Team, W, L, T columns

## Preprocessing

```python
df
```

## Create new dataframe with desired output

```python
df_nfl_data = pd.DataFrame(
    columns=["Year", "Teams", "Wins", "Losses", "Ties", "#Games"]
)
```

```python
grouped = df.groupby("Year")

for name, group in grouped:
    teams = group["Team"].tolist()
    wins = group["W"].tolist()
    losses = group["L"].tolist()
    ties = group["T"].tolist()

    ngames = wins[0] + losses[0] + ties[0]
    # concat to the dataframe

    df_tmp = (
        pd.Series(
            {
                "Year": name,
                "Teams": teams,
                "Wins": wins,
                "Losses": losses,
                "Ties": ties,
                "#Games": ngames,
            }
        )
        .to_frame()
        .T
    )
    df_nfl_data = pd.concat([df_nfl_data, df_tmp])
```

```python
# Save the data
df_nfl_data.to_parquet("../../prepared_data/NFL_data.parquet", index=False)
```
