```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Load Data

```python
df = pd.read_csv("../raw_data/baseballdata.csv")
df.head()
```

## Format of the data:

- Wins, Losses and Ties per Team per Season from 1876 to 2016
- We only need Year, Tm , W, L, Ties columns

## Preprocessing

## Create new dataframe with desired output

```python
df_mlb_data = pd.DataFrame(columns=["Year", "Teams", "Wins", "Losses", "Ties", "#Games"])
```

```python
grouped = df.groupby('Year')

for name, group in grouped:
    teams = group['Tm'].tolist()
    wins = group['W'].tolist()
    losses = group['L'].tolist()
    ties = group['Ties'].tolist()

    ngames = wins[0]+losses[0]+ties[0]
    # concat to the dataframe

    df_tmp = pd.Series({"Year": name, "Teams": teams, "Wins": wins, "Losses": losses, "Ties": ties, "#Games": ngames}).to_frame().T
    df_mlb_data = pd.concat([df_mlb_data ,df_tmp])
```

```python
# Save the data
df_mlb_data.to_parquet("../prepared_data/MLB_data.parquet", index=False)
```
